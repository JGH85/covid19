```python
import pandas as pd
```

```python
# url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv'
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
raw_data = pd.read_csv(url)
```

```python
raw_data.head()
```

```python
raw_data.columns
```

```python
covid_data = raw_data.rename(
columns = {'Province/State':'subregion', 'Country/Region':'country', 'Lat':'lat', 'Long':'long'})
```

```python
covid_data.head()
```

```python
# use melt to turn date columns into data rows
# id_vars is all the columns not converted
# var is the column that gets converted to a row
# val is the value from those columns
keep_columns = ['country', 'subregion', 'lat', 'long']
covid_data = (covid_data.melt(id_vars = keep_columns, var_name = 'date_raw', value_name = 'confirmed' ))
```

```python
covid_data
```

```python
# convert string dates into date time
covid_data.filter(['date_raw'])
```

```python
covid_data.assign(date = pd.to_datetime(covid_data.date_raw, format = '%m/%d/%y')).filter(['date', 'date_raw'])
```

```python
# group by date to remove duplicates, add confirmed column
(covid_data
.assign(date = pd.to_datetime(covid_data.date_raw, format = '%m/%d/%y'))
.filter(['date', 'date_raw', 'confirmed'])
.groupby(['date', 'date_raw'])
.agg('sum')
.sort_values('date')
)
```

```python
# make a backup before assigning date
covid_data_backup_before_date_conversion = covid_data.copy()
```

```python
#now just use the assign tested above but applied to main dataframe to add new date column
covid_data = covid_data.assign(date = pd.to_datetime(covid_data.date_raw, format = '%m/%d/%y'))
covid_data.head()
```

```python
# filter columns that we want
covid_data_final = covid_data.filter(['country', 'subregion', 'date', 'lat', 'long', 'confirmed'])
covid_data_final.head()
```

```python
# add a sort
covid_data_final = covid_data_final.sort_values(['country', 'subregion', 'lat', 'long', 'date'])
covid_data_final.head()
```

```python
# finally set index to country for now
covid_data_final.set_index('country', inplace = True)
covid_data_final.head()
```
